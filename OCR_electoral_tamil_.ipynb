```
#preprocessing of image files 

for k in range(4,count-2):


  B = 'input your image path '+str(k)+'.jpg'
  im = Image.open(B)
  im = im.resize((4000, 4882), PIL.Image.LANCZOS)
  im = cv2.bilateralFilter(np.array(im, dtype= np.uint8),9,75,75)
  im = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
  im = cv2.adaptiveThreshold(np.array(im,dtype=np.uint8), 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 31, 15)
  #segmentation of images after preprocessing 
  im = Image.fromarray(im)
  crop_rectangle_nm = (170,305, 1010, 610)
  for i in range(0,10):

    for j in range(0,3):
      
      Crop = im.crop(crop_rectangle_nm)
      Crop.save('page'+str(k)+'crop'+str(i)+str(j)+'.jpg')
      

      crop_rectangle_nm = tuple(np.add(crop_rectangle_nm,(1215,0,1215,0)))
    crop_rectangle_nm = tuple(np.subtract(crop_rectangle_nm,(3645,0,3645,0)))
    crop_rectangle_nm = tuple(np.add(crop_rectangle_nm,(0,405,0,410)))
```
